```python
X=data.iloc[:,1:5]
```

```python
#Normalisation
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)

```

```python

norm_func(X)
```

```python
dendrogram=sch.dendrogram(sch.linkage(norm_func(X),method='average'))
```

```python

#Create Cluster
model=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
model
```

```python

y=model.fit_predict(norm_func(X))
y
```

```python

data['h_clusterid']=model.labels_
data
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(data['h_clusterid'],data['Murder'], c=model.labels_)
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(norm_func(X))
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python

# selecting 4 clusters from above scree plot
model1=KMeans(n_clusters=4)
model1.fit(norm_func(X))
model1.labels_
```
